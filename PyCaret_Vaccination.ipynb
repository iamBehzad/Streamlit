```python
from pycaret.regression import *
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
train_df = pd.read_csv('country_vaccinations.csv')
#test = pd.read_csv('country_vaccinations_by_manufacturer.csv')


train_df['date'] = pd.to_datetime(train_df['date'], format = '%Y-%m-%d')

le = LabelEncoder()
#train_df['Day_Label'] = le.fit_transform(train_df.date)
train_df['Day'] = train_df['date'].dt.day
train_df['Month'] = train_df['date'].dt.month
train_df['Year'] = train_df['date'].dt.year


# Spliting target variable and independent variables
#train = train_df[['country',"Day_Label", "people_vaccinated_per_hundred"]]
train = train_df[['country','vaccines','Day','Month','Year', 'people_vaccinated_per_hundred']]

#le = LabelEncoder()

train['country'] = le.fit_transform(train['country'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)

train['vaccines'] = le.fit_transform(train['vaccines'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)

train['people_vaccinated_per_hundred'] = train['people_vaccinated_per_hundred'].fillna(0)
#train = StandardScaler().fit_transform(train)

train
```
